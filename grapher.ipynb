```python
import matplotlib.pyplot as plt
import json
```

### K vs Running Time

```python
# Read data
ks =[]
runningTimes =[]
outliers=50
kNN=30
with open('out/corel_k_to_time.jsonl') as f:
    while True:
        line = f.readline()
        if not line:
            break
        
        run = json.loads(line)
        runningTimes.append(run['runningTime'])
        ks.append(run['k'])
```

```python
# Plot
fig = plt.figure()
plt.plot(ks, runningTimes, marker='.', markersize=16, label='CorelColorMoments')
plt.xlabel('Number of input K to DHCA)', fontsize=14)
plt.ylabel('Running Time (s)', fontsize=14)
plt.title('RT (Top {}, {}nn) with input K to DHCA'.format(outliers, kNN))
plt.legend(loc='best')
plt.show()
# Save
fig.savefig('gen/corel_k_to_time.jpg')
```
